```python
from rdkit import Chem

# Ваш список SMILES строк
smiles_list = ['CC(C)(O)C(=O)O', 'CCC(O)(CC)C(=O)O', 'O=C(O)C1(O)CCCC1', 'O=C(O)C1(O)CCCCC1', 'O=C(O)C(O)c1ccccc1', 'CC(O)C(C)(O)C(=O)O', 'O=C(O)C(O)(CO)CO', 'O=C(O)[C@H](O)[C@@H](O)[C@H](O)[C@H](O)CO', 'O=C([O-])C1(O)C[C@@H](O)C(O)[C@H](O)C1']

# Функция для извлечения "обрывков" из SMILES строк
def extract_fragments(smiles_list):
    fragments = set()
    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            atoms = mol.GetAtoms()
            for atom in atoms:
                atom_symbol = atom.GetSymbol()
                fragments.add(atom_symbol)
            bonds = mol.GetBonds()
            for bond in bonds:
                bond_type = bond.GetBondType()
                fragments.add(str(bond_type))
    return list(fragments)

# Извлекаем "обрывки"
fragments_list = extract_fragments(smiles_list)

# Выводим результат
print(fragments_list)
```
