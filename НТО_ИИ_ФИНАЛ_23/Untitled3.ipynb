```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline

from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem, Draw
from rdkit.Chem.Draw import IPythonConsole

from lstm_chem.utils.config import process_config
from lstm_chem.model import LSTMChem
from lstm_chem.finetuner import LSTMChemFinetuner
from lstm_chem.data_loader import DataLoader
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/NTO/НТО_ИИ_ФИНАЛ_23/lstmchem (8) (1).csv')

# Extract the 'smiles' column as a list
smiles_list = ['O=C(O)CN(CC[SH](Cc1ccccn1)Cc1cnccn1)Cc1ccccn1'] + df['O=C(O)CN(CC[SH](Cc1ccccn1)Cc1cnccn1)Cc1ccccn1'].tolist()

# Write the list to a text file
with open('/content/LSTM_Chem/train.txt', 'w') as f:
    for smiles in smiles_list:
        f.write('%s\n' % smiles)
```

```python
config = process_config('experiments/2020-03-24/LSTM_Chem/config.json')

modeler = LSTMChem(config, session='finetune')
finetune_dl = DataLoader(config, data_type='finetune')

finetuner = LSTMChemFinetuner(modeler, finetune_dl)
finetuner.finetune()
```

```python
finetuned_smiles = finetuner.sample(num=400)
finetuned_smiles
```

```python
ans = list(set(finetuned_smiles))[0:100]
len(ans)
```

```python
import pandas as pd

df = pd.DataFrame(data={'smiles': ans})
df.to_csv('/content/lstmchem.csv', index=False, header=False)
```
